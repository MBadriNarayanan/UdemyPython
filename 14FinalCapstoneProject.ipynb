```python
import math

class Complex() :
    
    def __init__(self,r = 5,i = 5) :
        
        self.real = r
        
        self.imaginary = i
        
        if self.imaginary >= 0 :
            
            self.sign = '+'
            
        else :
            
            self.sign = '-'
        
    def __str__(self) :
        
        return '\n The Complex Number Is : {} {} {}i \n'.format(self.real,self.sign,abs(self.imaginary))
        
    def plus(self,b) :
        
        c = Complex()
        
        c.real = self.real + b.real
        
        c.imaginary = self.imaginary + b.imaginary 
        
        if c.imaginary >= 0 :
            
            c.sign = '+'
            
        else :
            
            c.sign = '-'
            
        return c
        
    def minus(self,b) :
        
        c = Complex()
        
        c.real = self.real - b.real
        
        c.imaginary = self.imaginary - b.imaginary 
        
        if c.imaginary >= 0 :
            
            c.sign = '+'
            
        else :
            
            c.sign = '-'
            
        return c
    
    def multiply(self,b) :
        
        c = Complex()
        
        c.real = (self.real * b.real) - (self.imaginary * b.imaginary) 
        
        c.imaginary = (self.real * b.imaginary) + (b.real * self.imaginary) 
        
        if c.imaginary >=0 :
            
            c.sign = '+'
            
        else :
            
            c.sign = '-'
            
        return c
    
    def divide(self,b) :
        
        c = Complex()
        
        c.real = ((self.real * b.real) + (b.imaginary * self.imaginary))/((self.real ** 2) + (b.real ** 2))
        
        c.imaginary = ((self.imaginary * b.real) - (self.real * b.imaginary))/((self.real ** 2) + (b.real ** 2))
        
        if c.imaginary >=0 :
            
            c.sign = '+'
            
        else :
            
            c.sign = '-'
            
        return c
        
```

```python
def get_real_imaginary() :
    
    real = int(input(' Enter Real Part : '))
    
    imaginary = int(input(' Enter Imaginary Part : '))
    
    return (real,imaginary)
```
